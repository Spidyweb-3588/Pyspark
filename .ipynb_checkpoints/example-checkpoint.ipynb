```python
import findspark
findspark.init()

from pyspark.sql import SparkSession
from pyspark.sql import Row

spark = SparkSession \
    .builder \
    .appName("sample") \
    .master("local[*]") \
    .getOrCreate()

sc = spark.sparkContext
```

```python
df = spark.read.format("parquet").option("header","true").option("inferSchema","true").load("C:/spark/spark-2.4.7-bin-hadoop2.7/examples/src/main/resources/users.parquet")
df.show(20, False)
df.printSchema()
print(type(df))

#주소값이 열이 되고, 나머지 도큐먼트(딕셔너리)형식(속에 키:값(리스트)의 값 하나로 매핑되어있다.
#너무많은 열(열이 URL로 되어있기 때문)이 생성되게 된다. 값인 도큐먼트의 형식은 일정한 형태를 가지고 있다.
```
